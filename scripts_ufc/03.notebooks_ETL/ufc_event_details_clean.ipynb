```python
# Import necessary libraries
import pandas as pd

# Load necessary data
df_events = pd.read_csv('../02.csv_scrapping/ufc_event_details.csv')

# Standardize column names
df_events.columns = df_events.columns.str.lower().str.strip().str.replace(' ', '_')

# Convert 'date' column to datetime format
df_events['date'] = pd.to_datetime(df_events['date'], errors='coerce')

# Filter out rows < 2016-01-01
df_events = df_events[df_events['date'] >= '2016-01-01'].reset_index(drop=True)

# Save cleaned data to a new CSV file
df_events.to_csv('../04.csv_clean/ufc_event_details_clean.csv', index=False)

# Display the first 5 rows of the cleaned DataFrame
df_events.head(5)   
```
