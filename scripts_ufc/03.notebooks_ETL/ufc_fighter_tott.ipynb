```python
# Import necessary libraries and read the CSV file
import pandas as pd
df_fighter_tott = pd.read_csv('../02.csv_scrapping/ufc_fighter_tott.csv')

# Standardize column names
df_fighter_tott.columns = df_fighter_tott.columns.str.lower().str.strip().str.rstrip('.').str.replace('. ', '.').str.replace('.', '_').str.replace(' ', '_')

# Display the first few rows of the DataFrame
df_fighter_tott.columns
```

```python
# Bucle for delete % and convert to float
for col in ['str_acc', 'str_def', 'td_acc', 'td_def']:
    df_fighter_tott[col] = df_fighter_tott[col].astype(str).str.replace('%', '', regex=True).str.strip()
    df_fighter_tott[col] = pd.to_numeric(df_fighter_tott[col], errors='coerce')
    df_fighter_tott[col] = df_fighter_tott[col] / 100

# Convert 'Weight' from pounds to kilograms
def weight_to_kg(weight):
    try:
        lbs = pd.to_numeric(weight.replace('lbs.', '').strip(), errors='coerce')
        kg = lbs * 0.453592
        return kg
    except:
        return None
    
# Apply the 'weight_to_kg' function to the 'Weight' column
df_fighter_tott['weight'] = df_fighter_tott['weight'].apply(weight_to_kg)

# Connvert 'Height' from feet and inches to centimeters
def height_to_cm(height):
    try:
        feet, inches = height.split("'")
        feet = int(feet.strip())
        inches = int(inches.replace('"', '').strip())
        total_inches = (feet * 12) + inches
        cm = total_inches * 2.54
        return cm
    except:
        return None

# Apply the 'height_to_cm' function to the 'Height' column
df_fighter_tott['height'] = df_fighter_tott['height'].apply(height_to_cm)


# Convert 'Reach' from feet and inches to centimeters
def reach_to_cm(reach):
    try:
        inches = pd.to_numeric(reach.replace('"', '').strip(), errors='coerce')
        cm = inches * 2.54
        return cm
    except:
        return None

# Apply the 'reach_to_cm' function to the 'Reach' column
df_fighter_tott['reach'] = df_fighter_tott['reach'].apply(reach_to_cm)

# Convert 'dob' to datetime format
df_fighter_tott['dob'] = pd.to_datetime(df_fighter_tott['dob'], errors='coerce')

# Rename columns for clarity
df_fighter_tott.rename(columns={'td_avg': 'td_avg_15m', 'sub_avg': 'sub_avg_15m', 'str_acc': 'str_acc_%', 'str_def': 'str_def_%', 'td_acc': 'td_acc_%', 'td_def': 'td_def_%'}, inplace=True)

# Save the cleaned dataframe to a new CSV file
df_fighter_tott.to_csv('../04.csv_clean/ufc_fighter_tott.csv', index=False)

# Filter 'Ilia Topuria' data
df_fighter_tott = df_fighter_tott[df_fighter_tott['fighter'] == 'Ilia Topuria'].copy()

# Display the cleaned dataframe
df_fighter_tott.head(5)
```
