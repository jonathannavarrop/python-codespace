```python
# Import library and dataframe
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load dataframe
df = pd.read_csv('ufc_fight_results_clean.csv')

# Filter by weight class    
weight_class = 'Flyweight Bout'  # example: 'lightweight', 'welterweight', 'middleweight', etc.
df = df[df['WEIGHTCLASS'].str.lower() == weight_class.lower()]

# Not a integer column
df = df['AGE.DIFF.WINNER'] == df['AGE.DIFF.WINNER'].astype(float)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Crear columna binaria: 1 si ganó el mayor, 0 si ganó el menor
df['older_wins'] = df['AGE.DIFF.WINNER'].apply(lambda x: 1 if x > 0 else 0)

# 2. Agrupar por cada diferencia de edad exacta y calcular % victorias
# (redondeando la diferencia a años enteros si fuera decimal)
df['age_diff_rounded'] = df['AGE.DIFF.WINNER'].round(0).astype(int)

win_percent = df.groupby('age_diff_rounded')['older_wins'].mean() * 100

# 3. Graficar
plt.figure(figsize=(10,5))
plt.plot(win_percent.index, win_percent.values, marker='o')
plt.xlabel('Age Difference (Winner - Loser in years)')
plt.ylabel('% Wins by Older Fighter')
plt.title('Victory Rate vs Age Difference in UFC')
plt.grid(True)
plt.show()
```
