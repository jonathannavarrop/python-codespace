```python
# Import necessary libraries
import pandas as pd
import datetime

# Load the dataset
df = pd.read_csv('ufc_fighter_tott.csv')
df.head(5)
```

```python
# Bucle for delete % and convert to float
for col in ['Str. Acc.', 'Str. Def', 'TD Acc', 'TD Def.']:
    df[col] = df[col].astype(str).str.replace('%', '', regex=True).str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col] / 100


# Calculate the age of each fighter and drop the 'DOB' column
current_date = pd.to_datetime('today')
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')
df['AGE'] = ((current_date - df['DOB']).dt.days) / 365
df = df.drop(columns=['DOB'])

# Connvert 'Height' from feet and inches to centimeters
def height_to_cm(height):
    try:
        feet, inches = height.split("'")
        feet = int(feet.strip())
        inches = int(inches.replace('"', '').strip())
        total_inches = (feet * 12) + inches
        cm = total_inches * 2.54
        return cm
    except:
        return None

# Apply the height_to_cm function to the 'Height' column
df['HEIGHT'] = df['HEIGHT'].apply(height_to_cm)

# Convert 'Reach' from feet and inches to centimeters
def reach_to_cm(reach):
    try:
        inches = pd.to_numeric(reach.replace('"', '').strip(), errors='coerce')
        cm = inches * 2.54
        return cm
    except:
        return None

# Apply the reach_to_cm function to the 'Reach' column
df['REACH'] = df['REACH'].apply(reach_to_cm)

# Sort dataframe
df = df[['FIGHTER','WEIGHT', 'AGE', 'HEIGHT',  'REACH', 'STANCE', 'SLpM', 'Str. Acc.', 'TD Avg', 'TD Acc', 'Sub. Avg.', 'SApM', 'Str. Def',  'TD Def.' ]]

df.rename(columns={'TD Avg': 'TD Avg 15min', 'Sub. Avg.': 'Sub. Avg 15min', 'Str. Acc.': 'Str. Acc %', 'Str. Def': 'Str. Def %', 'TD Acc': 'TD Acc %', 'TD Def.': 'TD Def. %'}, inplace=True)

# Save the cleaned dataframe to a new CSV file
df.to_csv('ufc_fighter_tott_clean.csv', index=False)

# Display the cleaned dataframe
df.head(5)
```
